```python
re_df = pd.read_csv(csvfile)
re_df.head()
```

```python
#Remove unnecessary columns

re_df_clean = re_df[["Month", "ZipCode", "ZipName", "Median Listing Price", "Active Listing Count ", "Days on Market ",]]
re_df_clean.head()
```

```python
#Break down 
df60611 = re_df_clean[re_df_clean["ZipCode"] == 60611.0]
df60611.head()
```

```python
df60062 = re_df_clean[re_df_clean["ZipCode"] == 60062.0]
df60062.head()
```

```python
df60523 = re_df_clean[re_df_clean["ZipCode"] == 60523.0]
df60523.head()
```

```python
df60118 = re_df_clean[re_df_clean["ZipCode"] == 60118.0]
df60118.head()
```

```python
#Visualize real estate pricing trends for each zip code

months60611 = df60611["Month"]
price60611 = df60611["Median Listing Price"]

months60062 = df60062["Month"]
price60062 = df60062["Median Listing Price"]

months60523 = df60523["Month"]
price60523 = df60523["Median Listing Price"]

months60118 = df60118["Month"]
price60118 = df60118["Median Listing Price"]


water_tower, = plt.plot(months60611, price60611, color="red", label="60611")
northbrook, = plt.plot(months60062, price60062, color="blue", label="60062")
oak_brook, = plt.plot(months60523, price60523, color="green", label="60523")
dundee, = plt.plot(months60118, price60118, color="orange", label="60118")

plt.title("Median Home Prices in Target Zip Codes")
plt.xlabel("May 2012 - Sept 2017")
plt.ylabel("Median Home Price")

plt.legend(handles=[water_tower, northbrook, oak_brook, dundee], loc="best")
```
