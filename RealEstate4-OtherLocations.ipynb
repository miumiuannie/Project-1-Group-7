```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Read in real estate data csv file
csvfile = 'Resources/RDC_InventoryCoreMetrics_Zip_IL.csv'
```

```python
re_df = pd.read_csv(csvfile)
re_df.head()
```

```python
#Remove unnecessary columns

re_df_clean = re_df[["Month", "ZipCode", "ZipName", "Median Listing Price", "Active Listing Count ", "Days on Market ",]]
re_df_clean.head()
```

```python
zips = pd.unique(re_df_clean["ZipCode"])
len(zips)
```

```python
#Sort by days on market
#short time on market suggest more people are moving into the area than out of it.  
#Pulled the top 100 out of 529 records

fast_sells = re_df_clean.sort_values("Days on Market ", ascending=True)
fastest_sells = fast_sells.head(100)
fastest_sells
```

```python
#From this list, we want to look at zip codes that showed good results in 2017 as opposed to 2012

#fast_sells_2017 = fastest_sells.sort_values("Month", ascending=False)
#fast_sells_2017
#that didn't work because of the date formatting, so let's separate out the year
fastest_sells["Year"] = fastest_sells["Month"].str[-4:]
fastest_sells

```

```python
#sort by year and see what comes up

fast_sells_2017 = fastest_sells.loc[fastest_sells["Year"] == "2017", :]
fast_sells_2017
```

```python
###### find the fastest sales in 2017

fast_sells_2017 = fastest_sells.loc[fastest_sells["Year"] == "2017", :]
fast_sells_2017

fastest_sells_2017 = fast_sells_2017.sort_values("Days on Market ", ascending=True)
fastest_sells_2017
```

```python
# Used Google Maps to select some target zip codes from this top 100 outside of the Chicago metro area:
#61752 (Le Roy), 61723 (Atlanta), 61462 (Monmouth), 62239 (Dupo), 62563 (Rochester), 61953 (Tuscola), 
#62230 (Breese), 61024 (Durand), 61073 (Roscoe)
```

```python
#Now, chart home pricing trends for each of the selected zip codes.

df61752 = re_df_clean[re_df_clean["ZipCode"] == 61752]
df61723 = re_df_clean[re_df_clean["ZipCode"] == 61723]
df61462 = re_df_clean[re_df_clean["ZipCode"] == 61462]
df62239 = re_df_clean[re_df_clean["ZipCode"] == 62239]
df62563 = re_df_clean[re_df_clean["ZipCode"] == 62563]
df61953 = re_df_clean[re_df_clean["ZipCode"] == 61953]
df62230 = re_df_clean[re_df_clean["ZipCode"] == 62230]
df61024 = re_df_clean[re_df_clean["ZipCode"] == 61024]
df61073 = re_df_clean[re_df_clean["ZipCode"] == 61073]

df61073.head()
```

```python
#Visualize real estate pricing trends for each zip code
#61752 (Le Roy), 61723 (Atlanta), 61462 (Monmouth), 62239 (Dupo), 62563 (Rochester), 61953 (Tuscola), 
#62230 (Breese), 61024 (Durand), 61073 (Roscoe)

months61752 = df61752["Month"]
price61752 = df61752["Median Listing Price"]
months61723 = df61723["Month"]
price61723 = df61723["Median Listing Price"]
months61462 = df61462["Month"]
price61462 = df61462["Median Listing Price"]
months62239 = df62239["Month"]
price62239 = df62239["Median Listing Price"]
months62563 = df62563["Month"]
price62563 = df62563["Median Listing Price"]
months61953 = df61953["Month"]
price61953 = df61953["Median Listing Price"]
months62230 = df62230["Month"]
price62230 = df62230["Median Listing Price"]
months61024 = df61024["Month"]
price61024 = df61024["Median Listing Price"]
months61073 = df61073["Month"]
price61073 = df61073["Median Listing Price"]

#61752 (Le Roy), 61723 (Atlanta), 61462 (Monmouth), 62239 (Dupo), 62563 (Rochester), 61953 (Tuscola), 
#62230 (Breese), 61024 (Durand), 61073 (Roscoe)

Le_Roy, = plt.plot(months61752, price61752, color="red", label="61752")
Atlanta, = plt.plot(months61723, price61723, color="blue", label="61723")
Monmouth, = plt.plot(months61462, price61462, color="green", label="61462")
Dupo, = plt.plot(months62239, price62239, color="orange", label="62239")
Rochester, = plt.plot(months62563, price62563, color="yellow", label="62563")
Tuscola, = plt.plot(months61953, price61953, color="cyan", label="61953")
Breese, = plt.plot(months62230, price62230, color="magenta", label="62230")
Durand, = plt.plot(months61024, price61024, color="burlywood", label="61024")
Roscoe, = plt.plot(months61073, price61073, color="blueviolet", label="61073")

plt.title("Median Home Price Trends in Proposed Zip Codes - Outside Chicagoland")
plt.xlabel("May 2012 - Sept 2017")
plt.ylabel("Median Home Price")

plt.legend(handles=[Le_Roy, Atlanta, Monmouth, Dupo, Rochester, Tuscola, Breese, Durand, Roscoe], loc="best")

plt.savefig("proposed_outside_trends.PNG")
```
