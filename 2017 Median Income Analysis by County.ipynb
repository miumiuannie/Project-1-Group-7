```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csvfile = 'ACS 5 YR -2017.csv'
```

```python

median_income_2017 = [50000, 55000, 60000, 65000, 70000, 75000, 80000, 85000, 90000]
x_axis = np.arange(len(median_income_2017))
```

```python
plt.bar(x_axis, median_income_2017, color='g', alpha=0.9, align="center")
```

```python
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Woodford", "Grundy", "Monroe", "Kane", "Will", "McHenry", "Lake", "DuPage", "Kendall"])
```

```python
plt.xlim(-0.15, len(x_axis)-0.75)
```

```python
plt.title("2017 Median Income")
plt.xlabel("Counties")
plt.ylabel("Income Per 20K")
```
