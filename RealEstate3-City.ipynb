```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Read in real estate data csv file
csvfile = 'Resources/RDC_InventoryCoreMetrics_Zip_IL.csv'
```

```python
re_df = pd.read_csv(csvfile)
re_df.head()
```

```python
#Remove unnecessary columns

re_df_clean = re_df[["Month", "ZipCode", "ZipName", "Median Listing Price", "Active Listing Count ", "Days on Market ",]]
re_df_clean.head()
```

```python
zips = pd.unique(re_df_clean["ZipCode"])
len(zips)
```

```python
#Sort by days on market
#short time on market suggest more people are moving into the area than out of it.  
#Pulled the top 100 out of 529 records

fast_sells = re_df_clean.sort_values("Days on Market ", ascending=True)
fastest_sells = fast_sells.head(100)
fastest_sells
```

```python
#From this list, we want to look at zip codes that showed good results in 2017 as opposed to 2012

#fast_sells_2017 = fastest_sells.sort_values("Month", ascending=False)
#fast_sells_2017
#that didn't work because of the date formatting, so let's separate out the year
fastest_sells["Year"] = fastest_sells["Month"].str[-4:]
fastest_sells

```

```python
#sort by year and see what comes up

fast_sells_2017 = fastest_sells.loc[fastest_sells["Year"] == "2017", :]
fast_sells_2017
```

```python
###### find the fastest sales in 2017

fast_sells_2017 = fastest_sells.loc[fastest_sells["Year"] == "2017", :]
fast_sells_2017

fastest_sells_2017 = fast_sells_2017.sort_values("Days on Market ", ascending=True)
fastest_sells_2017
```

```python
# Used Google Maps to select some target zip codes from this top 100 in the Chicago metro area, now
# 60606, 60622, 60661, 60642, 60603, 60601, 60607, 60605, 60612
```

```python
#Now, chart home pricing trends for each of the selected zip codes.

df60606 = re_df_clean[re_df_clean["ZipCode"] == 60606]
df60622 = re_df_clean[re_df_clean["ZipCode"] == 60622]
df60661 = re_df_clean[re_df_clean["ZipCode"] == 60661]
df60642 = re_df_clean[re_df_clean["ZipCode"] == 60642]
df60603 = re_df_clean[re_df_clean["ZipCode"] == 60603]
df60601 = re_df_clean[re_df_clean["ZipCode"] == 60601]
df60607 = re_df_clean[re_df_clean["ZipCode"] == 60607]
df60605 = re_df_clean[re_df_clean["ZipCode"] == 60605]
df60612 = re_df_clean[re_df_clean["ZipCode"] == 60612]

df60606.head()
```

```python
#Visualize real estate pricing trends for each zip code
# 60606, 60622, 60661, 60642, 60603, 60601, 60607, 60605, 60612

months60606 = df60606["Month"]
price60606 = df60606["Median Listing Price"]
months60622 = df60622["Month"]
price60622 = df60622["Median Listing Price"]
months60661 = df60661["Month"]
price60661 = df60661["Median Listing Price"]
months60642 = df60642["Month"]
price60642 = df60642["Median Listing Price"]
months60603 = df60603["Month"]
price60603 = df60603["Median Listing Price"]
months60601 = df60601["Month"]
price60601 = df60601["Median Listing Price"]
months60607 = df60607["Month"]
price60607 = df60607["Median Listing Price"]
months60605 = df60605["Month"]
price60605 = df60605["Median Listing Price"]
months60612 = df60612["Month"]
price60612 = df60612["Median Listing Price"]

# 60606, 60622, 60661, 60642, 60603, 60601, 60607, 60605, 60612

chi60606, = plt.plot(months60606, price60606, color="red", label="60606")
chi60622, = plt.plot(months60622, price60622, color="blue", label="60622")
chi60661, = plt.plot(months60661, price60661, color="green", label="60661")
chi60642, = plt.plot(months60642, price60642, color="orange", label="60642")
chi60603, = plt.plot(months60603, price60603, color="yellow", label="60603")
chi60601, = plt.plot(months60601, price60601, color="cyan", label="60601")
chi60607, = plt.plot(months60607, price60607, color="magenta", label="60607")
chi60605, = plt.plot(months60605, price60605, color="burlywood", label="60605")
chi60612, = plt.plot(months60612, price60612, color="blueviolet", label="60612")

plt.title("Median Home Price Trends in Proposed Zip Codes - Chicago")
plt.xlabel("May 2012 - Sept 2017")
plt.ylabel("Median Home Price")

plt.legend(handles=[chi60606, chi60622, chi60661, chi60642, chi60603, chi60601, chi60607, chi60605, chi60612], loc="best")

plt.savefig("proposed_city_trends.PNG")
```
